```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#Даалгавар 1.Интернетээс програмчлалын хэлний хэрэглээг 2022 оны байдлаар хайж олоод хэл тус бүрээр bar chart байгуул. Pyton, Java, Php, JavaScript, C#, C++, R гэх мэт
data = [12,10.12,1.5,2.3,6.4,7.73]
label = ['Python','Java','PHP','Javascript','C#','C++']
explode = (0,0.2,0.3,0.3,0,0)
plt.pie(data, explode=explode, labels=label, autopct='%1.1f%%',shadow=True, startangle=90)
plt.title(label='TIOBE:Popularity Of Programming Languages Worldwide, Dec 2022')
```

```python
#Даалгавар 2. Өмнө байгуулсан chart -аа хэвтээ хэлбэрээр байгуул
x = np.array(['Python','Java','PHP','Javascript','C#','C++'])
y = np.array([12,10.12,1.5,2.3,6.4,7.73])
plt.barh(x,y)
plt.title(label='TIOBE:Popularity Of Programming Languages Worldwide, Dec 2022')
plt.show()
```

```python
#Даалгавар 3. Өмнө байгуулсан chart -ийн багана бүрийн өнгийг өөрчил.
x = np.array(['Python','Java','PHP','Javascript','C#','C++'])
y = np.array([12,10.12,1.5,2.3,6.4,7.73])
plt.barh(x,y, color = 'Purple', height = 0.6)
plt.title(label='TIOBE:Popularity Of Programming Languages Worldwide, Dec 2022')
plt.show()
```
