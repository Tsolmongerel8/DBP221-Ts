```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# 1.    2022        bar chart . Pyton, Java, Php, JavaScript, C#, C++, R  
data = [12,10.12,1.5,2.3,6.4,7.73]
label = ['Python','Java','PHP','Javascript','C#','C++']
explode = (0,0.2,0.3,0.3,0,0)
plt.pie(data, explode=explode, labels=label, autopct='%1.1f%%',shadow=True, startangle=90)
plt.title(label='TIOBE:Popularity Of Programming Languages Worldwide, Dec 2022')
```

```python
# 2.   chart -   
x = np.array(['Python','Java','PHP','Javascript','C#','C++'])
y = np.array([12,10.12,1.5,2.3,6.4,7.73])
plt.barh(x,y)
plt.title(label='TIOBE:Popularity Of Programming Languages Worldwide, Dec 2022')
plt.show()
```
